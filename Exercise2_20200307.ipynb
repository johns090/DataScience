```python
import csv
import re


#mypath = os.getcwd()
phonePattern = '^\d{3}-\d{3}-\d{4}$'
emailPattern = '^\w+.?\w+@\w+.[a-z]{3}$'

goodRecords = []
badPhoneRecords = []
badEmailRecords = []

#try: 
#except:
#else:


with open('ca_sample_data_a.csv') as f:
    reader = csv.DictReader(f)     #New Ordered dictionary in version 3.6 

    for row in reader:
        #better check and ignore header
        #print(row)
        #print(dict(row)) 
        if not re.search(phonePattern, row['phone1']):
            badPhoneRecords.append(row)
        elif not re.search(emailPattern, row['email']):
            badEmailRecords.append(row)
        else:
            goodRecords.append(row)

#
#print(badPhoneRecords)
#print(badEmailRecords)
            
f.close()

fieldnames = ['first_name','last_name','company_name','address','city','province',
 'postal','phone1','email','web','exam_1','exam_2','exam_3']

# try catch for open files sestions
# can be done within Pandas instead of Python too

with open('bad_phone_recs.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    
    # can use writerows command without out loop
    for row in badPhoneRecords:
        #print(row)
        writer.writerow(row)
    
csvfile.close()

with open('bad_email_recs.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    
    # can use writerows command without out loop
    for row in badEmailRecords:
        writer.writerow(row)
    
csvfile.close()

with open('good_recs.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    
    # can use writerows command without out loop
    for row in goodRecords:
        writer.writerow(row)
    
csvfile.close()


        
        
        
```

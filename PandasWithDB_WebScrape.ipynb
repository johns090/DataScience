```python
import urllib.request as ur

# Search browser common headers
# Setup a request header, this makes your scraping more standard 
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'} 
reg_url = 'http://www.python.org'

# Request url with headers (Notice we are using a different method) 
req = ur.Request(url=reg_url, headers=headers)

# Read content of url 
html = ur.urlopen(req).read().decode()

# print the contents of the variable (containing the web page code)
print(html) 

```

```python
import urllib.request as ur 
from bs4 import BeautifulSoup

url = 'http://www.groupce.com/python/html/thejourney.html' 
response = ur.urlopen(url) 

#remember you can also use Request( with headers) 
html = response.read() # No need to decode() it if we are using BeautifulSoup

# The magic is here! 
parsedhtml = BeautifulSoup(html, "html.parser")

a_trs = parsedhtml('tr')
#print(a_trs)

sum = 0

for a_tr in a_trs:
    a_tds = a_tr('td')
    #print(a_tds[1].contents[0])
    number = int(a_tds[1].contents[0])
    sum += number

print(sum)
    
a_tags = parsedhtml('a')
for rows in a_tags: 
    print(rows) 
    print(rows['href']) 
    print(rows.contents[0]) 
    
```
